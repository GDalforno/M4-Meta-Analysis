```python
import pandas as pd
import numpy as np
```

```python
def change_index(df, index):
    df.index = df[index]
    df.drop(index, axis=1, inplace=True)
    return df
```

```python
def filter_submissions(df):
    df = change_index(df, "id")
    df = df.loc[m4_info.index]
    return df
```

```python
def smape(a, b):
    a = np.reshape(a, (-1,))
    b = np.reshape(b, (-1,))
    return np.mean(2.0 * np.abs(a - b) / (np.abs(a) + np.abs(b))).item()


def mase(insample, y_test, y_hat_test, freq):
    y_hat_naive = []
    for i in range(freq, len(insample)):
        y_hat_naive.append(insample[(i - freq)])

    masep = np.mean(abs(insample[freq:] - y_hat_naive))

    return np.mean(abs(y_test - y_hat_test)) / masep


def all_metrics(insample, y_test, y_hat_test, y_hat_naive, freq):
    m1 = smape(y_test, y_hat_test)
    m2 = mase(insample, y_test, y_hat_test, freq)
    
    m1_naive = smape(y_test, y_hat_naive)
    m2_naive = mase(insample, y_test, y_hat_naive, freq)
    
    owa = (m1/m1_naive + m2/m2_naive)/2
    return [m1, m2, owa]
```

```python
def store_metrics(data, index, name):
    data = pd.DataFrame(data, columns=["SMAPE", "MASE", "OWA"])
    data.index = index
    data.to_csv(f"./Metrics/{name}.csv")
```

```python
for freq, period_train, period_test in zip((1, 4, 12), (yearly_train, quarterly_train, monthly_train), (yearly_test, quarterly_test, monthly_test)):
    for index in period_train.index:
        train = period_train.loc[index,:].dropna().values
        test = period_test.loc[index,:].values

        naive2_z = naive2.loc[index,:].dropna().values
        
        trotta_metrics.append(all_metrics(train, test, trotta.loc[index,:].dropna().values, naive2_z, freq))
        alves_metrics.append(all_metrics(train, test, alves.loc[index,:].dropna().values, naive2_z, freq))
        muk_metrics.append(all_metrics(train, test, muk.loc[index,:].dropna().values, naive2_z, freq))
        rnn_metrics.append(all_metrics(train, test, rnn.loc[index,:].dropna().values, naive2_z, freq))
        pelka_metrics.append(all_metrics(train, test, pelka.loc[index,:].dropna().values, naive2_z, freq))
        mlp_metrics.append(all_metrics(train, test, mlp.loc[index,:].dropna().values, naive2_z, freq))
        theta_metrics.append(all_metrics(train, test, theta.loc[index,:].dropna().values, naive2_z, freq))
        arima_metrics.append(all_metrics(train, test, arima.loc[index,:].dropna().values, naive2_z, freq))
        damped_metrics.append(all_metrics(train, test, damped.loc[index,:].dropna().values, naive2_z, freq))
        ets_metrics.append(all_metrics(train, test, ets.loc[index,:].dropna().values, naive2_z, freq))
        holt_metrics.append(all_metrics(train, test, holt.loc[index,:].dropna().values, naive2_z, freq))
        ses_metrics.append(all_metrics(train, test, ses.loc[index,:].dropna().values, naive2_z, freq))
```
